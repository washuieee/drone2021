```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
images = glob.glob("*.png")
images = [cv2.imread(fname) for fname in images]
images = [cv2.pyrDown(image) for image in images]
```

```python
fig = plt.figure(figsize=(14, 6))
grid = ImageGrid(fig, 111,nrows_ncols=(2, 5),  axes_pad=0.1, )
for ax, im in zip(grid, images):
    ax.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```

# Red

```python
def process(image):
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    threshR = cv2.bitwise_or(
            cv2.inRange(hsv, (0, 150, 100), (10, 255, 255)),
            cv2.inRange(hsv, (160, 150, 100), (180, 255, 255)))
    return cv2.bitwise_and(rgb,rgb,mask=threshR)

fig = plt.figure(figsize=(14, 6))
grid = ImageGrid(fig, 111,nrows_ncols=(2, 5),  axes_pad=0.1, )
for ax, im in zip(grid, images):
    ax.imshow(process(im))
```
